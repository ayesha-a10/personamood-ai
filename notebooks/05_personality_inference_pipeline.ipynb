```python
import joblib
```

```python
tfidf = joblib.load(
    "../models/personality_model/tfidf.pkl"
)

traits = ["O", "C", "E", "A", "N"]

models = {}

for trait in traits:

    models[trait] = joblib.load(
        f"../models/personality_model/{trait}.pkl"
    )
```

```python
def predict_personality(text):

    X = tfidf.transform([text])

    predictions = {}

    for trait in traits:

        pred = models[trait].predict(X)[0]

        predictions[trait] = int(pred)

    return predictions
```

```python
def explain_personality(predictions):

    return {

        "Openness":
            "High" if predictions["O"] else "Low",

        "Conscientiousness":
            "High" if predictions["C"] else "Low",

        "Extraversion":
            "High" if predictions["E"] else "Low",

        "Agreeableness":
            "High" if predictions["A"] else "Low",

        "Neuroticism":
            "High" if predictions["N"] else "Low"
    }
```

```python
text = """
I enjoy meeting new people, attending events,
and trying new experiences.
"""

pred = predict_personality(text)

print(pred)

print(
    explain_personality(pred)
)
```

```python
def predict_personality_scores(text):

    X = tfidf.transform([text])

    results = {}

    for trait in traits:

        prob = models[trait].predict_proba(X)[0][1]

        results[trait] = round(
            float(prob),
            3
        )

    return results
```

```python
predict_personality_scores(
    text
)
```
