```python
import pandas as pd

train = pd.read_csv("../data/processed/train_processed.csv")
valid = pd.read_csv("../data/processed/validation_processed.csv")
test = pd.read_csv("../data/processed/test_processed.csv")
```

```python
X_train = train["clean_text"]
X_valid = valid["clean_text"]
X_test = test["clean_text"]

traits = ["O", "C", "E", "A", "N"]

y_train = train[traits]
y_valid = valid[traits]
y_test = test[traits]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=10000,
    ngram_range=(1,2),
    min_df=3
)

X_train_tfidf = tfidf.fit_transform(X_train)

X_valid_tfidf = tfidf.transform(X_valid)

X_test_tfidf = tfidf.transform(X_test)

print(X_train_tfidf.shape)
```

```python
from sklearn.linear_model import LogisticRegression

models = {}

for trait in traits:

    clf = LogisticRegression(
        max_iter=2000,
        random_state=42
    )

    clf.fit(
        X_train_tfidf,
        y_train[trait]
    )

    models[trait] = clf
```

```python
from sklearn.metrics import accuracy_score

for trait in traits:

    pred = models[trait].predict(
        X_test_tfidf
    )

    acc = accuracy_score(
        y_test[trait],
        pred
    )

    print(
        f"{trait}: {acc:.4f}"
    )
```

```python
import joblib
import os

os.makedirs(
    "../models/personality_model",
    exist_ok=True
)

for trait, model in models.items():

    joblib.dump(
        model,
        f"../models/personality_model/{trait}.pkl"
    )

joblib.dump(
    tfidf,
    "../models/personality_model/tfidf.pkl"
)
```
